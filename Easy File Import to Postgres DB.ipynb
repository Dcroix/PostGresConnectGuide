# Easy File Import to Postgres DB
This will work if Postgres DB server is running.
This uses Python programming. 

```python
# Import necessary libraries.
import pandas as pd
from sqlalchemy import create_engine

# Read the csv file for upload.
df = pd.read_csv("File.csv")
```

```python
# Create Engine
engine = create_engine("postgresql://user:password@localhost:5432/database_name")
# Import data to Postgres Database
df.to_sql("new_table_name", engine, if_exists="replace", index=False)
```
